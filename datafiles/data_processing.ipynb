```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"data.csv")

```

```python
df.describe(), df.info()
```

```python
df2 = df[df['mice_type']==1]
temp = df2.groupby(['bot_id', 'alpha'])['steps'].mean().reset_index()

plt.figure(figsize=(12, 8))

# line plot for each robot type
sns.lineplot(data=temp, x='alpha', y='steps', hue='bot_id', marker='o')
plt.title('One Stationary Mouse')
plt.xlabel('Alpha')
plt.ylabel('Steps')
plt.legend(title='Robot Type')

plt.show()
```

```python
df3 = df[df['mice_type']==2]
temp2 = df3.groupby(['bot_id', 'alpha'])['steps'].mean().reset_index()

plt.figure(figsize=(12, 8))

# line plot for each robot type
sns.lineplot(data=temp2, x='alpha', y='steps', hue='bot_id', marker='o')
plt.title('One Stochastic Mouse')
plt.xlabel('Alpha')
plt.ylabel('Steps')
plt.legend(title='Robot Type')

plt.show()
```
